```python
from fastai.data.transforms import get_image_files
from fastai.vision.augment import Resize, Brightness, Contrast, Saturation, Zoom, Rotate, RandomErasing
from fastai.vision.data import ImageDataLoaders

path = r'Original Image'


def is_rotten(x): return "Rotten" in x


augmentations = [
    Brightness(max_lighting=0.3, p=0.25),
    Contrast(max_lighting=0.2, p=0.25),
    Saturation(max_lighting=0.75, p=0.25),
    Zoom(max_zoom=1.3, p=0.25),
    Rotate(max_deg=90, p=0.3),
    RandomErasing(sh=0.30, max_count=1, p=1)
]

dls = ImageDataLoaders.from_name_func(
    path, get_image_files(path, recurse=True), valid_pct=0.2, seed=42,
    label_func=is_rotten, item_tfms=Resize(224), batch_tfms=augmentations)
```

```python
dls.show_batch()
```
